```python
import numpy as np
import pandas as pd


np.random.seed(42)
n=100
df=pd.DataFrame({
    
    "Size_sqft": np.random.randint(500, 3500, n),
    "Bedrooms": np.random.randint(1, 6, n),
    "Bathrooms": np.random.randint(1, 4, n),
    "Age_years": np.random.randint(1, 40, n),
    "Distance_city_km": np.random.randint(1, 25, n),
    "Floor": np.random.randint(1, 15, n)
})

df["Price"] = (
    df["Size_sqft"] * 120 +
    df["Bedrooms"] * 15000 +
    df["Bathrooms"] * 10000 -
    df["Age_years"] * 3000 -
    df["Distance_city_km"] * 2000 +
    df["Floor"] * 500 +
    np.random.randint(-20000, 20000, n)
)

x=df.drop("Price", axis=1)
y=df["Price"]
df
```
